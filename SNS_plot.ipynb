```
import seaborn as sns
import pandas as pd
```

```
import numpy as np
df = pd.DataFrame(np.random.randn(6000).reshape(3000,2), columns=["x1","x2"])
```

```
df.head()
```

```
sns.distplot(df['x1'])
```

```
sns.distplot(df['x2'])
```

```
sns.jointplot(df['x2'], df['x2'], kind='kde')
```

```
sns.pairplot(df)
```

```
days = ['sunday', 'monday', 'tuesday', 'wed', 'thurs', 'friday', 'saturday']
df['day'] = np.random.choice(7, 3000)
df['day'] = df.day.apply(lambda x: days[int(x)])
```

```
sns.catplot(x='day', y='x1', kind='box', data=df)
```

```
df.day.value_counts()
```
