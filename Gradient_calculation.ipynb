## Gradient computation using Tensorflow

```python
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
x = tf.linspace(-10.0,10.0,1000)
x = tf.Variable(x)
```

```python
## Sigmoid Gradient
with tf.GradientTape() as tape:
    y = 1/(1+tf.exp(-x))

dy = tape.gradient(y,x)
plt.plot(x.numpy() ,y.numpy(), color="red")
plt.plot(x.numpy(), dy.numpy(), color="blue")
plt.show()
```

```python
## Relu gradient
with tf.GradientTape() as tape:
    y = tf.maximum(0,x)

dy = tape.gradient(y,x)
plt.plot(x.numpy() ,y.numpy(), color="red")
plt.plot(x.numpy(), dy.numpy(), color="blue")
plt.show()
```

```python
## sin gradient
with tf.GradientTape() as tape:
    y = tf.sin(x)

dy = tape.gradient(y,x)
plt.plot(x.numpy() ,y.numpy(), color="red")
plt.plot(x.numpy(), dy.numpy(), color="blue")
plt.show()
```

```python
## tanh gradient
with tf.GradientTape() as tape:
    y = tf.tanh(x)

dy = tape.gradient(y,x)
plt.plot(x.numpy() ,y.numpy(), color="red")
plt.plot(x.numpy(), dy.numpy(), color="blue")
plt.show()
```

```python
## tanh gradient
with tf.GradientTape() as tape:
    y = tf.tanh(x)

dy = tape.gradient(y,x)
plt.plot(x.numpy() ,y.numpy(), color="red")
plt.plot(x.numpy(), dy.numpy(), color="blue")
plt.show()
```
